```python
certified,denied=[],[]
for i in conts:
    condi_1=(visa_dataset['continent']==i)
    condi_2=(visa_dataset['case_status']=='Certified')
    condi_3=(visa_dataset['case_status']=='Denied')
    count_certified=len(visa_dataset[condi_1 & condi_2])
    count_denied=len(visa_dataset[condi_1 & condi_3])
    certified.append(count_certified)
    denied.append(count_denied)
```

# Pearson-Correlation-Coefficient(r)

- r will vary from -1 to 1

- -100 to 100
- -1 to 1
- -1 to 0 ==> neg relation
- -0.5: both features 50% negatively correlated
-  0 to 1 : positively related
-  0.7: both features 70% positively correlated
- 0: constant - no relation
